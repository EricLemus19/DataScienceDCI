```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 15, 10000)
y = x + np.random.randn(10000)
```

```python
plt.plot(x,y, '.', alpha=0.3)
plt.xlim(0,15)
plt.ylim(0,15)
```

```python
x = np.linspace(0, 15, 100)
y = x + np.random.randn(100)*0.5
```

```python
fig , ax = plt.subplots(figsize=(5,5))
gs = fig.add_gridspec(3,3)

ax1 = fig.add_subplot(gs[0,:])
ax1.plot(x, y, 'o', color = 'orange', ms = 5, lw=2 , mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')
ax1.set_xticks([])
ax1.set_yticks([])

ax2 = fig.add_subplot(gs[1,0:2])
ax2.plot(x, y, '^-', color = 'red', ms = 5, lw=3, mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')
ax2.set_xticks([])
ax2.set_yticks([])

ax3 = fig.add_subplot(gs[1:,2])
ax3.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = fig.add_subplot(gs[2,0])
ax4.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$')
ax4.set_xticks([])
ax4.set_yticks([])

ax5 = fig.add_subplot(gs[2,1])
ax5.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3, label = '$3x^3$', alpha=0.3)
ax5.set_xticks([])
ax5.set_yticks([]);
```

```python
fig = plt.figure(figsize=(5,5))

ax1 = plt.axes([0.0, 0.0, 0.3, 0.3])
ax1.plot(x, y, 'o', color = 'orange', ms = 5, lw=2 , mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')
ax1.set_xticks([])
ax1.set_yticks([])

ax2 = plt.axes([0.3, 0.0, 0.3, 0.3])
ax2.plot(x, y, '^-', color = 'red', ms = 5, lw=3, mfc='yellow', mec ='blue', mew=2)
ax2.set_xticks([])
ax2.set_yticks([])

ax3 = plt.axes([0.6, 0.0, 0.3, 0.6])
ax3.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3)
ax3.set_xticks([])
ax3.set_yticks([])

ax4 = plt.axes([0.0, 0.3, 0.6, 0.3])
ax4.plot(x, 3*y, '*', color = 'deeppink', ms = 5, lw=3)
ax4.set_xticks([])
ax4.set_yticks([])

ax5 = plt.axes([0.0, 0.6, 0.9, 0.3])
ax5.plot(x, y, 'o', color = 'orange', ms = 5, lw=2 , mfc='lime', mec ='black', mew=2, ls =':')
ax5.set_xticks([])
ax5.set_yticks([]);
#ax5.axis('off')
```
