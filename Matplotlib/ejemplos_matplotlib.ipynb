### Método funcional

```python
import matplotlib.pyplot as plt
import numpy as np
# %matplotlib inline para jupyter en lugar de plt.show()
```

```python
x = np.linspace(5, 15, 10)
y = 4 * x**2
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Titulo')
```

* `marker` : estilo del simbolo ('+', 'o', '*', 's', ',', '.', '1', '2', '3', '4'...)
* `ms` : tamaño del punto 
* `lw` : ancho de la linea 
* `ls` : estilo de la linea (‘-‘, ‘--’, ‘-.’, ‘:’, ‘steps’)
* `mfc` : color del simbolo (sin incluir la orilla) 
* `mec` : color de la orilla del simbolo 
* `mew` : ancho de linea de la orilla del simbolo

```python
x1 = np.linspace(5, 15, 10)
x2 = np.linspace(5, 15, 20)

y1 = 4 * x1**2
y2 = x2**3

plt.xlabel('Eje X [m/s]') 
plt.ylabel('Eje Y [kg]')
plt.plot(x2, y2, marker = 'o', color = 'orange', ms = 10, lw=4 , 
         mfc='lime', mec ='black', mew=2, ls ='-', label = '$x^3$')
plt.plot(x1, y1, '^', color = 'red', ms = 15, lw=3, 
         mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')
plt.title('titulo de grafica')
plt.legend()
plt.grid() # malla
```

```python
x1 = np.linspace(5, 15, 10)
x2 = np.linspace(5, 15, 20)

y1 = 4 * x1**2
y2 = x2**3

plt.xlabel('Eje X [m/s]') 
plt.ylabel('Eje Y [kg]')
plt.plot(x1, y1, '^', color = 'red', ms = 15, 
         mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')

plt.plot(x2, y2, 'o', color = 'orange', ms = 10, lw=4 , 
         mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')

plt.plot(x2, 3*y2, '*--', color = 'deeppink', ms = 15, 
         lw=3, label = '$3x^3$')
plt.title('titulo de grafica')
plt.xlim(5,10) # limites del eje x
plt.ylim(0,1000) # limites del eje y
plt.legend()
```

### Multiples graficas en la misma figura

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x1, y1, 'or--') # More on color options later
plt.subplot(1,2,2)
plt.plot(y2, x2, 'g*-');
```

### Método orientado a objetos

```python
x = np.linspace(5, 15, 10)
y = 4 * x**2

# Se crea la figura (vacía).
fig = plt.figure()

# añadir ejes
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label') 
axes.set_ylabel('Set y Label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main 
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset

ax1.plot(x, y, 'b')
ax1.set_xlabel('eje X')
ax1.set_ylabel('eje Y')
ax1.set_title('titulo principal')

ax2.plot(y, x, 'r')
ax2.set_xlabel('eje X2')
ax2.set_ylabel('eje Y2')
ax2.set_title('titulo');
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titulo')
```

```python
fig, ax = plt.subplots()

ax.plot(x, y, 'r')
ax.set_xlabel(r'$v_{\alpha}$ [$\frac{m}{s}$]', fontsize = 20, color = 'b')
ax.set_ylabel(r'$v_{\beta}$ [kg]', fontsize = 25)
ax.set_title(r'$\alpha \beta^3$', fontsize = 30)

ax.axvline(8, ls='-.', lw=3, c='y')
ax.text(6, 200, 'algo')
ax.text(0.7, 0.9, 'coords relativas', transform=ax.transAxes)
```

```python
x = np.linspace(5, 15, 10)
y = 4 * x**2

fig, ax = plt.subplots()

ax.plot(x, y, 'r')
ax.set_xlabel(r'$v_{\alpha}$ [$\frac{m}{s}$]', fontsize = 20, color = 'b')
ax.set_ylabel(r'$v_{\beta}$ [kg]', fontsize = 25)
ax.set_title(r'$\alpha \beta^3$', fontsize = 30)

ax.axvline(8, ls='-.', lw=3, c='y')
ax.text(6, 200, 'algo')
ax.text(0.7, 0.9, 'coords relativas', transform=ax.transAxes)
ax.set_xticks([2,4,6,8,10])
ax.set_yticks(range(200,1000,200))

ax.set_xticklabels(['A', 'B', 'C', 'D', 'E'], fontsize=20)
```

```python
fig, ax = plt.subplots(figsize=(12,3))
ax.plot(x, y, 'r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('titulo')
```

```python
fig, ax = plt.subplots(2,2, figsize=(8,8))
fig.suptitle(r'$\alpha \beta^3$' , fontsize=20)

ax[0,0].set_xlabel('Eje Xa [m/s]', fontsize=20)
ax[0,0].plot(x2, y2, 'o', color = 'orange', ms = 10, lw=4 , mfc='lime', mec ='black', mew=2, ls =':', label = '$x^3$')
ax[0,0].plot(x1, y1, '^-', color = 'red', ms = 15, lw=3, mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')
ax[0,0].legend()
ax[0,0].grid()

ax[0,1].set_xlabel('Eje Xb [m/s]', fontsize=20)
ax[0,1].plot(x2, y2, '^-', color = 'red', ms = 15, lw=3, mfc='yellow', mec ='blue', mew=2, label = '$4x^2$')

ax[1,0].set_xlabel('Eje Xc [m/s]', fontsize=20)
ax[1,0].plot(x2, 3*y2, '*', color = 'deeppink', ms = 15, lw=3, label = '$3x^3$', alpha=0.3)

ax[1,1].set_xlabel('Eje Xd [m/s]', fontsize=20)
ax[1,1].plot(x2, 3*y2, '*', color = 'deeppink', ms = 15, lw=3, label = '$3x^3$')
ax[1,1].legend()
ax[1,1].grid()

fig.subplots_adjust(hspace = 0.5, wspace = 0.3) # hspace -> espacio horizontal.   wspace -> espacio vertical
```

### Graficos de dispersión

```python
x = np.arange(100)
y = x * (np.random.randn(100))
```

```python
plt.plot(x,y, 'o')
```

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
```

```python
x = np.arange(100)
y = x * (np.random.randn(100))
z = np.random.randint(5,10,100) 

fig, ax = plt.subplots()
ax.scatter(x, y, c = z, cmap='viridis') # c= color
```

```python
fig, ax = plt.subplots()
im = ax.scatter(x, y, c = z, cmap='cool') # c= color
fig.colorbar(im, ax=ax)
```

```python
w = np.linspace(200,1, 100)

fig, ax = plt.subplots()
im = ax.scatter(x,y, c = z, s = w, cmap='plasma', alpha = 0.5) # s = tamaño del simbolo
fig.colorbar(im, ax=ax, label = 'variable color')
```
