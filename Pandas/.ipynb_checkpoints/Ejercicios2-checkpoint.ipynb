```python
df_mex = df[df['Country'] == 'Mexico']
df_mex.info()
```

```python
df_mex = df_mex.drop('Other',axis=1)
df_mex.info()
```

```python
df_mex.isna().sum(axis=0)
```

```python
#df_mex_clean = df_mex.dropna(axis=0, how='any') 
```

```python
df_global = df[df['Country'] == 'Global']
df_global.info()
```

```python
df_global = df_global.drop('Other',axis=1)
```

```python
mex_global = pd.merge(df_mex, df_global, left_on='Year', right_on='Year').sort_index()
```

```python
mex_global.columns
```

```python
mex_global.drop(['Country_x','ISO 3166-1 alpha-3_x', 'Country_y', 'ISO 3166-1 alpha-3_y'],axis=1, inplace=True)

new_names = {'Total_x':'Total_mex', 'Coal_x':'Coal_mex',
             'Oil_x':'Oil_mex', 'Gas_x':'Gas_mex', 
             'Cement_x':'Cement_mex', 'Flaring_x': 
             'Flaring_mex', 'Per Capita_x':'Per Capita_mex',
             'Total_y':'Total_glob', 'Coal_y':'Coal_glob',
             'Oil_y':'Oil_glob', 'Gas_y':'Gas_glob', 
             'Cement_y':'Cement_glob', 'Flaring_y': 
             'Flaring_glob', 'Per Capita_y':'Per Capita_glob'
}
mex_global = mex_global.rename(columns= new_names)
```

```python
mex_global.head()
```

```python
mex_global.info()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(mex_global['Year'], mex_global['Total_mex'], c='lime', lw=3, label='Mexico')
ax.plot(mex_global['Year'], mex_global['Total_glob'], c='b', lw=3, label='Global')
```

```python
fig, ax = plt.subplots(figsize=(6, 5))
ax.plot(mex_global['Year'], mex_global['Total_mex'], c='lime', lw=3, label='Mexico')
ax.plot(mex_global['Year'], mex_global['Total_glob'], c='b', lw=3, label='Global')
#ax.set_yscale('log')
```

```python
mex_global.columns
```

```python
cols_to_plot = ['Total_', 'Coal_', 'Oil_', 'Gas_', 'Cement_',
       'Flaring_', 'Per Capita_']

fig, ax = plt.subplots(7,1, figsize=(4, 20))
for i,col in enumerate(cols_to_plot):
    ax[i].plot(mex_global['Year'], mex_global[col+'mex'], c='lime', lw=3, label='Mexico')
    ax[i].plot(mex_global['Year'], mex_global[col+'glob'], c='b', lw=3, label='Global')
    ax[i].set_title(col[0:-1])

fig.subplots_adjust(hspace=0.4)
```
