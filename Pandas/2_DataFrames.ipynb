## **DataFrame:** 

Estructura (objeto) multivariable con indexaci√≥n integrada. _Tabla_

Conjunto de Series

```python
import pandas as pd
import numpy as np
```

```python
valores = [['Ana', 18, 0.5], ['Luis', 22, 0.33], ['Carmen', 20, 0.85]]
columnas = ['Nombre', 'Edad', 'Otro']
filas = ['a', 'b', 'c']

df = pd.DataFrame(data=valores, index=filas, columns=columnas)
df
```

```python
nums = np.random.randint(100, 3000, (20, 4))
letters = np.reshape(list("abcd")*5, (20,1))

values = np.append(nums, letters, axis = 1)
values
```

[pandas.date_range](https://pandas.pydata.org/docs/reference/api/pandas.date_range.html)

```python
fecha = pd.date_range("20230915", periods=20)
```

```python
df = pd.DataFrame(values, index=fecha, columns=list("VWXYZ"))
df
```

### Filtrando elementos

```python
df[['V', 'W']] > 1000
```

```python
df[df['W'] > 1000]
```

```python
df[df['W'] > 1000][['X', 'V']]
```

```python
df[df['W'] > 0][['Y','X']]
```

```python
df[(df['W'] <= 1500) & (df['V'] > 1000)]
```

```python
df2 = df[(df['W'] <= 1500) & (df['V'] > 1000)]

df2
```
