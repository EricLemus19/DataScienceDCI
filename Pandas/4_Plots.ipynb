```python
import pandas as pd
import numpy as np
```

```python
n = 100
a = np.random.randn(n,2)*300 + 1000
b = nums = np.random.randint(100, 3000, (n, 2))
c = np.append(a, b, axis=1)
```

```python
#nums = np.random.randint(100, 3000, (20, 4))
nums = np.random.randn(n,4)
letters = np.reshape(list("abcd")*int(n/4), (n,1))
values = np.append(c, letters, axis = 1)
fecha = pd.date_range("20230915", periods=n)

df = pd.DataFrame(values, index=fecha, columns=list("VWXYZ"))
```

```python
types_dict = {}
for i in df.columns[0:-1]:
    types_dict.update({i : 'float64'})

df = df.astype(types_dict)

df.head()
```

```python
df.dtypes
```

```python
df.plot()
```

```python
df.plot(figsize=(10,5), legend=False, title= 'Random int', xlabel = 'date')
```

```python
#import matplotlib.pyplot as plt
fig, ax0 = plt.subplots(figsize=(10,3))
df.plot(ax = ax0)
ax0.set_title('Random int')
ax0.get_legend().remove()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,3))
df[['X', 'Y']].plot(ax = ax[0])
ax[0].set_title('Random int')
ax[0].get_legend().remove()

df[['V', 'W']].plot(ax = ax[1])
ax[1].set_title('Random normal')
ax[1].get_legend().remove()
```

```python
df.groupby('Z').sum()
```

```python
df.groupby('Z').sum().plot.bar()
```
