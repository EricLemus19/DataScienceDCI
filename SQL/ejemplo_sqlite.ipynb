```python
import pandas as pd 
import sqlite3
```

```python
path = '../../'
database = path + 'database.sqlite'
```

```python
# iniciar coneccion con la base de datos
conn = sqlite3.connect(database)
```

```python
# estructura 'global' de la base de datos
tables = pd.read_sql("""SELECT * FROM sqlite_master
                        WHERE type='table';""", conn)
```

### Seleccionar columnas de una tabla

> SELECT col FROM tabla;

> SELECT * FROM tabla; 

\* -> todas

```python
countries = pd.read_sql("""SELECT *
                        FROM Country;""", conn)
countries
```

```python
league = pd.read_sql("""SELECT *
                        FROM League;""", conn)
league
```

```python
Player_Attributes = pd.read_sql("""SELECT *
                        FROM Player_Attributes
                        LIMIT 5;""", conn)

Player_Attributes
```

```python
player = pd.read_sql("""SELECT *
                        FROM Player
                        LIMIT 5;""", conn)
player
```

```python
match = pd.read_sql("""SELECT *
                        FROM Match
                        LIMIT 5;""", conn)
match
```

```python
team = pd.read_sql("""SELECT *
                        FROM Team
                        LIMIT 5;""", conn)
team
```

### ORDER BY

```python
pd.read_sql("""SELECT player_name, height, weight
                FROM Player
                ORDER BY height ASC
                LIMIT 5;""", conn)
```

```python
pd.read_sql("""SELECT player_name, weight
                FROM Player
                ORDER BY height DESC
                LIMIT 5;""", conn)
```

### WHERE

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height <= 180
                LIMIT 5;""", conn)
```

```python
pd.read_sql("""SELECT player_name, weight, height
                FROM Player
                WHERE height BETWEEN 180 AND 190
                LIMIT 5;""", conn)
```
